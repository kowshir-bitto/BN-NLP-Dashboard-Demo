```python
import re
import emoji

def contains_bangla(text):
    bangla_regex = re.compile(r'[\u0980-\u09FF]')
    return bool(bangla_regex.search(text))

def remove_emojis(text):
    return emoji.replace_emoji(text, replace=" ")

def clean_bangla_text(text):
    stop_words = set(stop_list)
    punctuations = set(punc_list)
    text = remove_emojis(text)
    text = re.sub(r"[^\u0980-\u09FF\s]", "", text)
    words = text.split()
    cleaned_words = [word for word in words if word not in stop_words and word not in punctuations]
    return " ".join(cleaned_words)

def classify_bangla_sentiment(text, positive_emotions, negative_emotions):
    words = set(text.split())
    if words & positive_emotions:
        return 'positive'
    elif words & negative_emotions:
        return 'negative'
    return 'neutral'
```

```python
from collections import Counter

all_words = ' '.join(df['Cleaned_Comment']).split()
word_counts = Counter(all_words)
top_15_words = word_counts.most_common(15)

words, counts = zip(*top_15_words)

fig3 = go.Figure(data=[go.Bar(
    x=words,
    y=counts,
    marker_color='purple'
)])
fig3.update_layout(title='Top 15 Words in Cleaned Comments')
fig3.show()
Top15WordsinCleanedComments:"PlotlyChart" = fig3
```
