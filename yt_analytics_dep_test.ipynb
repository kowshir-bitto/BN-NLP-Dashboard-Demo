##General Cleaning Functions

```python
def contains_bangla(text):
    bangla_regex = re.compile(r'[\u0980-\u09FF]')
    return bool(bangla_regex.search(text))

def remove_emojis(text):
    return emoji.replace_emoji(text, replace=" ")

def clean_bangla_text(text):
    stop_words = set(stop_list)
    punctuations = set(punc_list)
    text = remove_emojis(text)
    text = re.sub(r"[^\u0980-\u09FF\s]", "", text)
    words = text.split()
    cleaned_words = [word for word in words if word not in stop_words and word not in punctuations]
    return " ".join(cleaned_words)

def classify_bangla_sentiment(text):
    words = set(text.split())
    if words & positive_emotions:
        return 'positive'
    elif words & negative_emotions:
        return 'negative'
    return 'neutral'
```

##Live Youtube Comments Sentiment Analytics

###Data Crawling using API

```python
def get_video_id(url):
    pattern = r"(?:v=|v\/|embed\/|youtu.be\/|\/v\/|watch\?v=|&v=)([A-Za-z0-9_-]{11})"
    match = re.search(pattern, url)
    if match:
        return match.group(1)
    return None

def extract_youtube_comments(url,api_key):
    video_id = get_video_id(url)
    if not video_id:
        raise ValueError("Invalid YouTube URL or unable to extract video ID.")

    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=api_key)
    comments = []
    try:
        video_response = youtube.videos().list(id=video_id, part="snippet").execute()
        if not video_response["items"]:
            raise ValueError("No video found with the given ID.")
        video_title = video_response["items"][0]["snippet"]["title"]

        next_page_token = None
        while True:
            comment_response = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                pageToken=next_page_token,
                maxResults=100
            ).execute()

            for item in comment_response["items"]:
                comment = item["snippet"]["topLevelComment"]["snippet"]
                user_id = comment["authorDisplayName"]
                comment_text = comment["textDisplay"]

                if contains_bangla(comment_text):
                    comments.append({"video_title": video_title, "user_id": user_id, "comment": comment_text})

            next_page_token = comment_response.get("nextPageToken")
            if not next_page_token:
                break
    except Exception as e:
        print(f"An error occurred: {e}")

    return pd.DataFrame(comments)
```

```python
def visualize_word_frequency_ytv(df):
    all_words = " ".join(df["cleaned_comment"])
    words = all_words.split()
    word_freq = pd.Series(words).value_counts().reset_index()
    word_freq.columns = ["word", "count"]

    fig = px.bar(
        word_freq.head(20),
        x="word",
        y="count",
        title="Top 20 Words in Cleaned Comments",
        labels={"word": "Word", "count": "Frequency"}
    )
    return fig
```

##Sentiment Proportion

```python
def visualize_emotion_distribution_ytc(df):
    emotion_counts = df["emotion"].value_counts().reset_index()
    emotion_counts.columns = ["emotion", "count"]

    fig = px.bar(
        emotion_counts,
        x="emotion",
        y="count",
        title="Emotion Distribution of Bangla Comments",
        labels={"emotion": "Emotion", "count": "Count"},
        color="emotion",
        color_discrete_map={'(?)':'black', 'positive':'green', 'negative':'red', 'neutral':'blue'}
    )
    return fig

#visualize_emotion_distribution_ytc(df_2_x).show()
```

##Sentiment Sunburst

```python
def visualize_sunburst_with_ytcomments(df):
    df["count"] = 1

    fig = px.sunburst(
        df,
        path=["emotion", "user_id", "comment"],
        values="count",
        title="Public Sentiment for Youtube Video",
        color="emotion",
        color_discrete_map={'(?)':'black', 'positive':'green', 'negative':'red', 'neutral':'blue'}
    )
    return fig


#visualize_sunburst_with_ytcomments(df_2_x).show()
```
